```python
#Obtiene el path con el nombre de todos los archivos de un directorio.
def get_files_from_path(directory):
    path_files = []
    dir_list = os.listdir(directory)
    for path in dir_list:
        path_files.append(directory+"\\"+path)
    return path_files


#Extrae los paths que cumplan con un codigo
def extract_paths_for_emotions_keys(emotions_code, files_path, get_code):
    paths = []
    emotions_set = set(emotions_code)
    for code_file in files_path:
        if (get_code(code_file) in emotions_set):
            paths.append(code_file)
    return paths


#Obtiene el codigo en el nombre del archivo para el dataset CREMA-D
def get_code_esp(path):
    return path[114]#path[110] prueba

#Esta funciÃ³n abre el archivo .wav y obtiene el mfcc escalado en un vector de 40 elementos.
def duplicate(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    augmented_signal = pitch_scale(audio, sample_rate,2)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_pitchscale{file_extension}", augmented_signal , sample_rate)

def polarity(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    polarity_signal = audio * (-1)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_polarity{file_extension}", polarity_signal , sample_rate) 
    
def time_stretch(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    time_stretch_signal = librosa.effects.time_stretch(audio, 0.5)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_time_stretch{file_extension}", time_stretch_signal , sample_rate) 
    
def random_gain(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    gain_factor = random.uniform(2, 8)
    signal_with_gain = audio * gain_factor
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_random_gain{file_extension}", signal_with_gain  , sample_rate)

#Selecciona n ejemplos que necesitemos y los mezcla.
def select_elements(examples, code, quantity, new_code):
    random.shuffle(examples)
    print(len(examples))
    elements = []
    counter = 1
    for example in examples:
        if (counter > quantity):
            break
        if code == example[1]:
            elements.append((example[0],new_code))
            counter = counter + 1
    print(counter, len(elements))
    return elements

def pitch_scale(signal,sr,num_semitones):
    return librosa.effects.pitch_shift(signal, sr, num_semitones)
```
