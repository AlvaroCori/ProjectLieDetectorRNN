```python
import librosa
import librosa.display
import os
import soundfile as sf
import random
random.seed(0)
```

```python
#Obtiene el path con el nombre de todos los archivos de un directorio.
def get_files_from_path(directory):
    path_files = []
    dir_list = os.listdir(directory)
    for path in dir_list:
        path_files.append(directory+"\\"+path)
    return path_files

#Esta funci√≥n abre el archivo .wav y obtiene el mfcc escalado en un vector de 40 elementos.
def duplicate(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    augmented_signal = pitch_scale(audio, sample_rate,2)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_pitchscale{file_extension}", augmented_signal , sample_rate)
    
#Estas funciones son diferentes formas de modificar el audio, para mas informacion ver https://www.youtube.com/watch?v=umAXGVzVvwQ
def polarity(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    polarity_signal = audio * (-1)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_polarity{file_extension}", polarity_signal , sample_rate) 
    
def time_stretch(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    time_stretch_signal = librosa.effects.time_stretch(audio, 0.5)
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_time_stretch{file_extension}", time_stretch_signal , sample_rate) 
    
def random_gain(file_name):
    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    gain_factor = int(random.uniform(2, 7))
    signal_with_gain = audio * gain_factor
    filename, file_extension = os.path.splitext(file_name)
    sf.write(f"{filename}_random_gain_{gain_factor}{file_extension}", signal_with_gain  , sample_rate)

def pitch_scale(signal,sr,num_semitones):
    return librosa.effects.pitch_shift(signal, sr, num_semitones)
```

#### Obtenemos los audios a partir de la direccion actual, subiendo 2 niveles y metiendonos a las carpetas Datasets y data
